## Water Jug Problem using DFS:

```python

jug1_cap = 4
jug2_cap = 3
TARGET = 2

visited = set()

def dfs(state, path):
    x, y = state  
    if state in visited:
        return False
    # Mark this state visited
    visited.add(state)
    # Add current state to path
    path.append(state)
    
    if x == TARGET or y == TARGET:
        return True
    
    next_states = []
    
    # 1. Fill Jug1
    next_states.append((jug1_cap, y))
    # 2. Fill Jug2
    next_states.append((x, jug2_cap))
    # 3. Empty Jug1
    next_states.append((0, y))
    # 4. Empty Jug2
    next_states.append((x, 0))
    # 5. Pour Jug1 -> Jug2
    pour = min(x, jug2_cap - y)
    next_states.append((x - pour, y + pour))
    # 6. Pour Jug2 -> Jug1
    pour = min(y, jug1_cap - x)
    next_states.append((x + pour, y - pour))
    
    for new_state in next_states:
        if dfs(new_state, path):
            return True
    
    # Backtrack if no solution
    path.pop()
    return False

initial_state = (0, 0)
solution_path = []

found = dfs(initial_state, solution_path)

if found:
    print("Solution found!\n")
    print("Steps followed (state representation: (Jug1, Jug2)):\n")
    
    for step in solution_path:
        print(step)
else:
    print("No solution found.")
```
